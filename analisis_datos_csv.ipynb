```python
def analisis_eda (df):
    nombre_columnas = df["Age"]

    lista_diccionarios = []

    for col in nombre_columnas:

        dicc = {}

        dicc["nombre_columna"] = col

        dicc["unique"] = df[col].unique()

        # dicc["values_count"] = df[col].values_count()

        # dicc["dtype"] = df[col].select_dtypes()

        dicc["describe"] = df[col].describe()

        dicc["max"] = df[col].max()

        dicc["min"] = df[col].min()

        dicc["isnull"] = df[col].isnull()

        dicc["notnull"] = df[col].notnull()

        dicc["duplicated"] = df[col].duplicated()

        lista_diccionarios.append(dicc)

    df_analisis = pd.DataFrame(lista_diccionarios)

prueba = analisis_eda(df) 

prueba.head()
```
