```python
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)

# Leer el archivo CSV
df = pd.read_csv('HR RAW DATA.csv')


pd.set_option('display.max_colwidth', None)

```

### Limpieza de las últimas 14 columnas del Dataframe 'HR RAW DATA'

```python
#Copia de seguridad
ultimas_catorce = df.copy()
ultimas_catorce.head(3)
```

```python
#Funcion para cambiar ',' por '.' y tipo float

def cambiarcomafloat(df,columnas):
    for col in columnas:
        df[col] = df[col].str.replace(',', '.', regex = True).astype(float)
    return df

```

```python
#Dejar en dataframe solo las 14 ultimas columnas
ultimas_catorce.drop(['Age', 'Attrition', 'BusinessTravel', 'DailyRate',
       'Department', 'DistanceFromHome', 'Education', 'EducationField',
       'employeecount', 'employeenumber', 'EnvironmentSatisfaction', 'Gender',
       'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobRole',
       'JobSatisfaction', 'MaritalStatus', 'MonthlyIncome', 'MonthlyRate',
       'NUMCOMPANIESWORKED', 'Over18', 'OverTime', 'PercentSalaryHike',
       'PerformanceRating', 'RelationshipSatisfaction', 'StandardHours'], axis = 1, inplace = True)

#Visualizar el dataframe modificado
ultimas_catorce.head()

#Homegeneizar los nombre de las columnas para que queden en minuscula
columnas_limpias = {columna: columna.lower() for columna in ultimas_catorce.columns}
columnas_limpias

ultimas_catorce.rename(columns = columnas_limpias, inplace = True)

#Copia de seguridad del dataframe
ultimas_catorce1 = ultimas_catorce.copy()

#Eliminar columnas que no usaremos en el analisis
ultimas_catorce1.drop(['sameasmonthlyincome', 'salary', 'numberchildren'],  axis = 1, inplace = True)

#Crear copia de seguridad del dataframe
ultimas_catorce3 = ultimas_catorce1.copy()

#Llamar a funcion para cambiar ',' por '.'"HR RAW DATA.csv"columnas_cambiar = ['totalworkingyears', 'yearsincurrentrole', 'worklifebalance']
columnas_cambiar = ['totalworkingyears', 'yearsincurrentrole', 'worklifebalance']
ultimas_catorce3 =cambiarcomafloat(ultimas_catorce3, columnas_cambiar)

#Reemplazar los True por Si y los False por No
diccionario_mapa = {'True': "Si", 'False': "No", 'Yes' : 'Si', '0' : 'No', '1' : 'Si' }
ultimas_catorce3["remotework"] = ultimas_catorce3["remotework"].map(diccionario_mapa)

#Pasar todos los datos de la columna roledepartment a minuscula
ultimas_catorce3["roledepartament"] = ultimas_catorce3["roledepartament"].str.lower()

#Crear copia de seguridad del dataframe
ultimas_catorce4 = ultimas_catorce3.copy()
ultimas_catorce5 = ultimas_catorce4.copy()

#Imputacion columna categorica roledepartament por Unknown
ultimas_catorce5['roledepartament'] = ultimas_catorce5['roledepartament'].fillna('unknown')
ultimas_catorce5['roledepartament'].isnull().sum()

#ultimas_catorce5.to_csv('ultimas_catorce.csv')
```

### Union de los tres archivos .csv generados después de la limpieza

```python
# Lectura de los 3 archivos .csv generados despues de la limpieza 
df1 = pd.read_csv('columnas_ana_primeras.csv', index_col = 0)
df2 = pd.read_csv('columnas_maria_2.csv', index_col = 0)
df3 = pd.read_csv('ultimas_catorce.csv', index_col = 0)
```

```python
#Eliminar la columna unnamed que nos crea Pandas por defecto
df2 = df2.drop(columns=['unnamed: 0'])

#Union del dataframe df1 con el df2
df1df2 = df1.merge(df2, left_index= True, right_index = True)
df1df2.head(3)
```

```python
#Eliminar la columna unnamed que nos crea Pandas por defecto
df3 = df3.drop(columns=['unnamed: 0'])
```

```python
#Union de dataframe anterior con el df3
limpieza_completo = df1df2.merge(df3, left_index= True, right_index = True)
limpieza_completo.head(3)
```
